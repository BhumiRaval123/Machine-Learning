```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
def binaryStep(x):
  return np.heaviside(x,1)

x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.title('binaryStep')
plt.show()
```

```python
def linear(x):
  return x
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.title('Linear')
plt.show()
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Sigmoid')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Tanh')
plt.show()
```

```python
def leaky_relu(x):
    if x<0:
        return 0.01*x
    else:
        return x
x=[float(x)/100 for x in range(-1000,1000)]
y = [leaky_relu(x1) for x1 in x]

plt.plot(x,y)
plt.title("leaky relu")
plt.show()
```

```python
def step(x):
    if x<0:
        return 0
    else:
        return 4


x=[float(x)/100 for x in range(-700,700)]
y = [step(x1) for x1 in x]
plt.plot(x,y)
plt.title("step")
plt.show()

```
